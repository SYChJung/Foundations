## 1. How many power plants are in each state?

```python
USpowerplant = pd.read_csv("/14-homework-powerplants/powerplants.csv")
USpowerplant['geometry'] = USpowerplant.apply(lambda row: Point(row.Longitude, row.Latitude), axis=1)
USpowerplant = gpd.GeoDataFrame(USpowerplant)
USpowerplant.crs = {'init':'epsg:4326'}
USpowerplant
```

```python
USpowerplant.dtypes
```

```python
USshape = gpd.read_file("/14-homework-powerplants/cb_2016_us_state_500k/cb_2016_us_state_500k.shp")
USshape
```

```python
USpowerplant.crs # {'init': 'epsg:4326'}
USshape.crs      # {'init': 'epsg:4269'}
```

```python
USshape = USshape.to_crs(USpowerplant.crs)
USshape.crs
```

```python
USshape['powerplant_count'] = USshape.apply(
    lambda row: USpowerplant.within(row.geometry).sum(), 
    axis=1
)
USshape.head()
```

## 3. Which state has the most hydroelectric power plants?

## 13. Which states have the most renewable plants, as a % of total plants? (include hydro, wind, solar)

```python
USpowerplant_sol = USpowerplant[~USpowerplant['Solar_MW'].isnull()]
USpowerplant_sol = USpowerplant_sol[ USpowerplant_sol['Solar_MW'] != 0.0]
USpowerplant_sol.shape
```

```python
USshape_renew = USshape.copy()
USshape_renew['Solar powerplant count'] = USshape_renew.apply(
    lambda row: USpowerplant_sol.within(row.geometry).sum(), axis=1)
USshape_renew.head(1)
```

```python
USpowerplant_wind = USpowerplant[~USpowerplant['Wind_MW'].isnull()]
USpowerplant_wind = USpowerplant_wind[ USpowerplant_wind['Wind_MW'] != 0.0]
USpowerplant_wind.shape
```

```python
USshape_renew['Wind powerplant count'] = USshape_renew.apply(
    lambda row: USpowerplant_wind.within(row.geometry).sum(), axis=1)
USshape_renew.head(1)
```

```python
USpowerplant_hydro = USpowerplant[~USpowerplant['Hydro_MW'].isnull()]
USpowerplant_hydro = USpowerplant_hydro[ USpowerplant_hydro['Hydro_MW'] != 0.0]
USpowerplant_hydro.shape
```

```python
USshape_renew['Hydro powerplant count'] = USshape_renew.apply(
    lambda row: USpowerplant_hydro.within(row.geometry).sum(), axis=1)
USshape_renew.head()
```

```python
USshape_renew['Renewable powerplant count'] = USshape_renew['Solar powerplant count'] + USshape_renew['Wind powerplant count'] + USshape_renew['Hydro powerplant count']
USshape_renew.head()
```

```python
s = USshape_renew['Solar powerplant count']
w = USshape_renew['Wind powerplant count']
h = USshape_renew['Hydro powerplant count']
t = USshape_renew['powerplant_count']

USshape_renew['percentage'] = round( (((s+w+h)/t) * 100), 2 )
USshape_renew.head()
```

```python
USshape_renew = USshape_renew.sort_values(["percentage"], ascending = False)
USshape_renew.reset_index()['NAME'][0]
```

## Preparation: Knowing your tags

These questions are the same for every data set, and might not work exactly for yours.

```python
from bs4 import BeautifulSoup
import requests
import requests as rq
```

```python
data = {
    "pht_status":'BAR',
    "pht_lic":'',
    "pht_lnm":'',
    "pht_fnm":'',
    "pht_oth_name":'',
    "phy_city":'HOUSTON',           
    "phy_cnty":'-1',
    "phy_zip":'',
    "B1":'Search'
}

headers = {
    "Referer": 'https://www.tdlr.texas.gov/cimsfo/'
}

response = rq.post("https://www.tdlr.texas.gov/cimsfo/fosearch_results.asp", data = data, headers = headers)
# response
doc = BeautifulSoup(response.text, "html.parser")
# doc
len(doc)
```

### What is the tag and class name for every row of data?

```python
# The tag is 'tr'
TBV_rows = doc.find_all('tr')[1:]
# TBV_rows
for tag in TBV_rows :
    print(tag.text.strip())
    
```

### What is the tag and class name for the violation number?

```python
# 'span', 'class': 'results_text'
TBV_rows = doc.find_all('tr')[1:]
for complaint_no in TBV_rows :
    if complaint_no :
        print(complaint_no.find_all('span', attrs = {'class': 'results_text'})[-2].text)
# 'company' is the problem
```

```python
# td
TBV_rows = doc.find_all('tr')[1:]
for description in TBV_rows :
    if len(description) :
        print(description.find_all('td')[2].text)
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number

Create a new dictionary for each `tr` (except the header).

```python
Texas_Barber_Violations = []

for element in TBV_rows[:-1]:
    TBV_dictionary = {}
    person_name = element.find_all('span')[0].text
    if person_name != "Company:":
#         print(person_name)
        TBV_dictionary['Person name'] = person_name
    description = element.find_all('td')[2].text
    if description :
#         print(description)
        TBV_dictionary['Violation description'] = description
    complaint_no = element.find_all('span', attrs = {'class': 'results_text'})[-2].text
    if complaint_no :
#         print(complaint_no)
        TBV_dictionary['Violation number'] = complaint_no 
    Texas_Barber_Violations.append(TBV_dictionary)
#     print(TBV_dictionary)
#     print("-----")
Texas_Barber_Violations

```

### Save that to a CSV

```python
import pandas as pd
df = pd.DataFrame(Texas_Barber_Violations)
# df.head()
df.to_csv("Texas_Barber_Violations.csv", index=False)
```

# Measuring International Bitterness Units

IBU stands for [International Bitterness Unit](http://www.thebrewenthusiast.com/ibus/), and while a lot of places like to brag about having the most bitter beer (it's an American thing!), IBUs don't necessary *mean anything*.

Let's look at how different beers have different IBU measurements.

## What is the average IBU measurement across all beers?

```python
print("The average IBU of the beers is", round(df['IBUs'].mean(), 2 ))
```

## Draw the distribution of IBU measurements, but with *thirty* bins instead of the default of 10

```python
df['IBUs'].hist(bins=30)
```

## List all of the beers with IBUs above the 75th percentile

```python
# print( df.shape )
# print( df[~df['IBUs'].isin( [ 'NaN' ] )].shape )
# print( df[~df['IBUs'].isnull()].shape )
# print( df[~df['IBUs'].isnull()].quantile( 0.75 ) )
percentile75 = df['IBUs'].quantile( 0.75 )
print(percentile75)
df[ percentile75 < df['IBUs'] ]
```

## List all of the beers with IBUs below the 25th percentile

```python
percentile25 = df['IBUs'].quantile( 0.25 )
print(percentile25)
df[ df['IBUs'] < percentile25 ]
```

## List the median IBUs of each type of beer. Graph it.

Put the highest at the top, and the nan ones at the bottom. I want a nice graph, too - don't let them all be squished together, either!

```python
# df.groupby('gender')['networthusbillion'].mean()
beer_IBU = df.groupby('Style')['IBUs'].median()
# beer_IBU.shape
beer_IBU.sort_values(ascending = False)
```

```python
beer_IBU.sort_values(ascending = True).plot(kind='barh', figsize=(40,40))
```

## Hmmmm, it looks like they are generally different styles. What are the most common 5 styles of high-IBU beer vs. low-IBU beer?

You'll want to think about it in three pieces - finding the high IBU beers, then finding out what the most common styles are, then getting the top 5.

```python
df['IBUs'].describe()
```

```python
# I defined "high-IBU beer" as beer styles that have IBUs over the 75 percentile,
# and "low-IBU beer" as beer styles that have IBUs under the 25 percentile
IBUs_described = df['IBUs'].describe()
df[ df['IBUs'] > IBUs_described.loc['75%'] ]
# IBUs_described.loc['75%']
high_IBU_beer = df[ IBUs_described.loc['75%'] < df['IBUs'] ]['Style'].value_counts().head()
print("The the most common 5 styles of high IBU beer are \n", high_IBU_beer)
```

```python
df[ df['IBUs'] < IBUs_described.loc['25%'] ]
# IBUs_described.loc['25%']
low_IBU_beer = df[ df['IBUs'] < IBUs_described.loc['25%'] ]['Style'].value_counts().head()
print("The the most common 5 styles of low IBU beer are \n", low_IBU_beer)
```

## Get the average IBU of "Witbier", "Hefeweizen" and "American Pale Wheat Ale" styles (also draw a histogram)

I want ONE measurement and ONE graph, not three separate ones. And 20 to 30 bins in the histogram, please.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### one mean 

```python
WHA = df[ df['Style'].isin(['Witbier', 'Hefeweizen', 'American Pale Wheat Ale']) ]
WHA_IBUs = WHA['IBUs']
print( round( WHA_IBUs.mean(), 2 )  )
WHA_IBUs.hist(bins=30)
```

#### three means

```python
WHA_group = WHA.groupby(['Style'])['IBUs']
print(WHA_group.mean())
```

```python
print(WHA_group.max())
print(WHA_group.min())
```

```python
fig_Witbier    = WHA[WHA['Style'] == 'Witbier']['IBUs'].dropna()
fig_Hefeweizen = WHA[WHA['Style'] == 'Hefeweizen']['IBUs'].dropna()
fig_APWA       = WHA[WHA['Style'] == 'American Pale Wheat Ale']['IBUs'].dropna()



fig, ax = plt.subplots()
for a in [fig_Witbier, fig_Hefeweizen, fig_APWA]:
    sns.distplot(a, bins=range(0, 66, 3), ax=ax, kde=False)
ax.set_xlim([0, 66])
```
